## Observations
Weight -> Price is inversely Proportional to Weight , With Increase in Weight Price Decreses weight between 100 to 200 have more Price

Resolution -> Increase in Resolution Increases the price till Resoltion is 6 ( Resolution between 4 to 6) Price is More

PPI -> Increase in PPI Increases the Price (400 to 600) more Price

CPU_core -> Increase in Core Increases the Price (if CPU core is 4 and 8 Price is More)

CPU_freq -> Price Increase  as Cpu_freq Increases ( above > 1)

internal Memory -> Price Increases as Internal Memeory Increases (120GB MB has Max Price) -> having Linear Relationship

RAM -> Price Increases as RAM Increases (6GB RAM  has Max Price) -> having Linear Relationship 

Rear_Camera -> With Increase in Rear Camera Price Increase (below  15 ) then after Price Reduced to some small Amount

Front_Camera -> with Increase in Front Camera Price Increases (below 7.5) then after Price Reduced to Some small Amount

Battery  -> with Increase in Battery Price Increases 

Thickness -> With Increase in thickness Price Decreases (thickness above  12)  price Reduced
