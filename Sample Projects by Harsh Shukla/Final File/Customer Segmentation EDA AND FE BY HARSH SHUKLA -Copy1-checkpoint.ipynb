## Binning Age , Annual_Income and Spending_Score

I have taken Reference from Kaggle

#### Age group 

```python
Age_18_25 = df.Age[(df.Age>=18) & (df.Age<=25)]
Age_26_35 = df.Age[(df.Age>=26) & (df.Age<=35)]
Age_36_45 = df.Age[(df.Age>=36) & (df.Age<=45)]
Age_46_55 = df.Age[(df.Age>=46) & (df.Age<=55)]
Age_55_Above = df.Age[(df.Age>=56)]
```

#### plotting of Age group vs No of Customers


```python
x = ['18-25' , '26-35','36-45','46-55','55 Above']
y = [len(Age_18_25.values) , len(Age_26_35.values) , len(Age_36_45.values), len(Age_46_55.values) , len(Age_55_Above.values)]
sns.barplot(x = x , y = y , palette="nipy_spectral_r")
plt.xlabel("Age" , fontsize = 12)
plt.ylabel("No of Customers" , fontsize = 12)
plt.title("Age Group vs No of Customers" , fontsize = 20 , weight = 'bold')
```

#### Spending Score Grouping 

```python
Spending_Score_1_20 = df.Spending_Score[(df.Spending_Score>=1) & (df.Spending_Score<=20)]
Spending_Score_21_40 = df.Spending_Score[(df.Spending_Score>=21) & (df.Spending_Score<=40)]
Spending_Score_41_60 = df.Spending_Score[(df.Spending_Score>=41) & (df.Spending_Score<=60)]
Spending_Score_61_80 = df.Spending_Score[(df.Spending_Score>=61) & (df.Spending_Score<=80)]
Spending_Score_81_100 = df.Spending_Score[(df.Spending_Score>=81) & (df.Spending_Score<=100)]
```

#### plotting of Spending_Score group vs No of Customers

```python
x = ['1-20' , '21-40','41-60','60-81','81-100']
y = [len(Spending_Score_1_20.values) , len(Spending_Score_21_40.values) , len(Spending_Score_41_60.values), len(Spending_Score_61_80.values) , len(Spending_Score_81_100.values)]
sns.barplot(x = x , y = y , palette="nipy_spectral_r")
plt.xlabel("Spending_Score (1 - 100)" , fontsize = 12)
plt.ylabel("No of Customers" , fontsize = 12)
plt.title("Spending_Score Group vs No of Customers" , fontsize = 20 , weight = 'bold')
```

#### Annual_Income Grouping

```python
ai0_30 = df["Annual_Income"][(df["Annual_Income"] >= 0) & (df["Annual_Income"] <= 30)]
ai31_60 = df["Annual_Income"][(df["Annual_Income"] >= 31) & (df["Annual_Income"] <= 60)]
ai61_90 = df["Annual_Income"][(df["Annual_Income"] >= 61) & (df["Annual_Income"] <= 90)]
ai91_120 = df["Annual_Income"][(df["Annual_Income"] >= 91) & (df["Annual_Income"] <= 120)]
ai121_150 = df["Annual_Income"][(df["Annual_Income"] >= 121) & (df["Annual_Income"] <= 150)]
```

#### plotting of Annual_Income group vs No of Customers

```python
x = ["$0 - 30,000", "$30,001 - 60,000", "$60,001 - 90,000", "$90,001 - 120,000", "$120,001 - 150,000"]
y = [len(ai0_30.values), len(ai31_60.values), len(ai61_90.values), len(ai91_120.values), len(ai121_150.values)]

sns.barplot(x = x , y = y, palette="nipy_spectral_r")
plt.xlabel("Annual_Income" , fontsize = 12)
plt.ylabel("No of Customers" , fontsize = 12)
plt.title("Annual_Income Group vs No of Customers" , fontsize = 20 , weight = 'bold')
```
