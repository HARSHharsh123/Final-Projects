```python
#Now Data Preprocessing (Feature Engieering )
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['gender'].value_counts()
```

```python
df['gender'] = df['gender'].map({'female' : 1 , 'male' : 0})
```

```python
df
```

```python
df['race_ethnicity'] = df['race_ethnicity'].map({'group A' : 1 , 'group B' : 2  , 'group C' : 3 , 'group D' : 4 , 
                                                'group E' : 5}) 
```

```python
df
```

```python
df.drop('parental_level_of_education' , axis =1, inplace = True)
```

```python
df
```

```python
df['lunch'] = df['lunch'].map({'standard' : 1 , 'free/reduced' : 0})
```

```python
df
```

```python
df['test_preparation_course'] = df['test_preparation_course'].map({'completed' : 1 , 'none' : 0})
```

```python
df
```

```python
df.drop(['math_score' ,'reading_score','writing_score'] , axis = 1 , inplace = True)
```
